```python
import numpy as np
import sympy as sp
```

Uma função e a sua derivada utilizando sympy

```python
fx = sp.symbols('x')**2 + 2*sp.symbols('x') + 1
flinhax = sp.Derivative(fx)
flinhax
```

```python
flinhax.doit()
```

Gráfico da Função e sua derivada

```python
sp.plot(fx, flinhax.doit())
```
